#### ðŸ§¬ What Are Mutations?

#### Types of Point Mutations:
1. **Silent (Synonymous)** - DNA changes but protein stays the same
2. **Missense (Nonsynonymous)** - DNA change causes different amino acid
3. **Nonsense** - DNA change creates stop codon

#### Example:
```
Original: ATG GCC TAA
          M   A   STOP

Silent:   ATG GCT TAA  (GCCâ†’GCT both code for A)
          M   A   STOP

Missense: ATG GTC TAA  (GCCâ†’GTC: Aâ†’V)
          M   V   STOP
```

#### ðŸŽ² Setting Up for Mutations

````python
import random

# Codon table for translation
CODON_TABLE = {
    "ATG": "M", "TTT": "F", "TTC": "F", "TTA": "L", "TTG": "L",
    "CTT": "L", "CTC": "L", "CTA": "L", "CTG": "L",
    "ATT": "I", "ATC": "I", "ATA": "I",
    "GTT": "V", "GTC": "V", "GTA": "V", "GTG": "V",
    "TCT": "S", "TCC": "S", "TCA": "S", "TCG": "S",
    "CCT": "P", "CCC": "P", "CCA": "P", "CCG": "P",
    "ACT": "T", "ACC": "T", "ACA": "T", "ACG": "T",
    "GCT": "A", "GCC": "A", "GCA": "A", "GCG": "A",
    "TAT": "Y", "TAC": "Y",
    "CAT": "H", "CAC": "H",
    "CAA": "Q", "CAG": "Q",
    "AAT": "N", "AAC": "N",
    "AAA": "K", "AAG": "K",
    "GAT": "D", "GAC": "D",
    "GAA": "E", "GAG": "E",
    "TGT": "C", "TGC": "C",
    "TGG": "W",
    "CGT": "R", "CGC": "R", "CGA": "R", "CGG": "R",
    "AGT": "S", "AGC": "S",
    "AGA": "R", "AGG": "R",
    "GGT": "G", "GGC": "G", "GGA": "G", "GGG": "G",
    "TAA": "*", "TAG": "*", "TGA": "*"  # Stop codons
}

def translate(dna):
    """Translate DNA to protein."""
    protein = []
    for i in range(0, len(dna), 3):
        codon = dna[i:i+3]
        if len(codon) < 3:
            break
        aa = CODON_TABLE.get(codon, "?")
        if aa == "*":
            break
        protein.append(aa)
    return "".join(protein)

print("Setup complete! âœ“")
````

#### ðŸ”§ Basic Mutation Function

````python
BASES = ["A", "C", "G", "T"]

def mutate(seq, n):
    """
    Introduce n random point mutations.
    
    Args:
        seq (str): DNA sequence
        n (int): Number of mutations
    
    Returns:
        tuple: (mutated_sequence, list_of_changes)
    """
    seq_list = list(seq)
    changes = []
    
    for _ in range(n):
        # Random position
        idx = random.randrange(len(seq_list))
        old = seq_list[idx]
        
        # Choose different base
        new_options = [b for b in BASES if b != old]
        new = random.choice(new_options)
        
        seq_list[idx] = new
        changes.append((idx, old, new))
    
    return "".join(seq_list), changes

# Test it
original = "ATGCCGATTGCA"
mutated, changes = mutate(original, 2)

print(f"Original: {original}")
print(f"Mutated:  {mutated}")
print(f"\nChanges:")
for pos, old, new in changes:
    print(f"  Position {pos}: {old} â†’ {new}")
````

#### ðŸ§¬ Analyzing Protein Impact

````python
def analyze_mutation_impact(original_dna, mutated_dna):
    """
    Analyze how mutations affect the protein.
    """
    # Translate both
    original_protein = translate(original_dna)
    mutated_protein = translate(mutated_dna)
    
    print("Mutation Impact Analysis")
    print("=" * 60)
    print(f"\nOriginal DNA: {original_dna}")
    print(f"Mutated DNA:  {mutated_dna}")
    
    # Show DNA differences
    dna_diffs = sum(1 for a, b in zip(original_dna, mutated_dna) if a != b)
    print(f"\nDNA changes: {dna_diffs}")
    
    # Show where DNA changed
    print("\nDNA Alignment:")
    markers = "".join("|" if a == b else "X" for a, b in zip(original_dna, mutated_dna))
    print(f"  {original_dna}")
    print(f"  {markers}")
    print(f"  {mutated_dna}")
    
    # Protein comparison
    print(f"\nOriginal Protein: {original_protein}")
    print(f"Mutated Protein:  {mutated_protein}")
    
    # Count amino acid changes
    if len(original_protein) == len(mutated_protein):
        aa_diffs = sum(1 for a, b in zip(original_protein, mutated_protein) if a != b)
        print(f"\nAmino acid changes: {aa_diffs}")
        
        # Classify mutations
        if aa_diffs == 0:
            print("Classification: All mutations are SILENT")
        else:
            print(f"Classification: {aa_diffs} NONSYNONYMOUS, {dna_diffs - aa_diffs} SILENT")
    else:
        print(f"\nâš ï¸  Length changed! Original: {len(original_protein)} aa, Mutated: {len(mutated_protein)} aa")
        print("   (Possibly introduced or removed stop codon)")

# Test it
original = "ATGCCGATTGCATAG"
mutated, _ = mutate(original, 3)
analyze_mutation_impact(original, mutated)
````

#### ðŸ“Š Silent vs Nonsynonymous Classification

````python
def classify_mutations(original_dna, mutated_dna):
    """
    Classify each mutation as silent or nonsynonymous.
    """
    # Find DNA changes
    dna_changes = []
    for i, (old, new) in enumerate(zip(original_dna, mutated_dna)):
        if old != new:
            dna_changes.append((i, old, new))
    
    if not dna_changes:
        print("No mutations found!")
        return
    
    print(f"Analyzing {len(dna_changes)} mutation(s)\n")
    
    # Translate
    original_protein = translate(original_dna)
    mutated_protein = translate(mutated_dna)
    
    silent_count = 0
    nonsynonymous_count = 0
    
    for pos, old_base, new_base in dna_changes:
        # Find which codon this is in
        codon_idx = pos // 3
        codon_pos = pos % 3
        
        # Get original and mutated codons
        codon_start = codon_idx * 3
        original_codon = original_dna[codon_start:codon_start+3]
        mutated_codon = mutated_dna[codon_start:codon_start+3]
        
        # Get amino acids
        if codon_idx < len(original_protein):
            original_aa = original_protein[codon_idx]
            mutated_aa = mutated_protein[codon_idx] if codon_idx < len(mutated_protein) else "?"
            
            # Classify
            if original_aa == mutated_aa:
                classification = "SILENT"
                silent_count += 1
            else:
                classification = "NONSYNONYMOUS"
                nonsynonymous_count += 1
            
            print(f"Position {pos} (Codon {codon_idx}, base {codon_pos}):")
            print(f"  {old_base} â†’ {new_base}")
            print(f"  Codon: {original_codon} â†’ {mutated_codon}")
            print(f"  Amino acid: {original_aa} â†’ {mutated_aa}")
            print(f"  Classification: {classification}\n")
    
    print("Summary:")
    print(f"  Silent: {silent_count}")
    print(f"  Nonsynonymous: {nonsynonymous_count}")
    print(f"  Total: {len(dna_changes)}")

# Test it
original = "ATGCCGATTGCATAG"
mutated, _ = mutate(original, 4)
classify_mutations(original, mutated)
````

#### ðŸŽ² Mutation Accumulation Over Time

````python
def simulate_evolution(seq, generations, mutations_per_gen):
    """
    Simulate mutation accumulation over multiple generations.
    """
    print(f"Simulating {generations} generations")
    print(f"Mutations per generation: {mutations_per_gen}\n")
    
    current_seq = seq
    original_protein = translate(seq)
    
    print(f"Generation 0 (original):")
    print(f"  DNA: {current_seq}")
    print(f"  Protein: {original_protein}\n")
    
    for gen in range(1, generations + 1):
        current_seq, changes = mutate(current_seq, mutations_per_gen)
        current_protein = translate(current_seq)
        
        # Calculate divergence
        dna_diff = sum(1 for a, b in zip(seq, current_seq) if a != b)
        protein_diff = sum(1 for a, b in zip(original_protein, current_protein) if a != b)
        
        print(f"Generation {gen}:")
        print(f"  DNA: {current_seq}")
        print(f"  Protein: {current_protein}")
        print(f"  DNA divergence: {dna_diff}/{len(seq)} ({dna_diff/len(seq)*100:.1f}%)")
        print(f"  Protein divergence: {protein_diff}/{len(original_protein)} ({protein_diff/len(original_protein)*100:.1f}%)\n")

# Simulate 5 generations
original = "ATGCCGATTGCATAG"
simulate_evolution(original, generations=5, mutations_per_gen=1)
````

#### ðŸŽ¯ Practice Exercise: Mutation Rate Comparison

````python
# Compare mutation rates
def test_mutation_rates():
    """
    Test how different mutation rates affect protein sequence.
    """
    original = "ATGCCGATTGCAGCGGGCTAG"
    rates = [1, 2, 5, 10]
    
    print(f"Original: {original}")
    print(f"Protein:  {translate(original)}\n")
    print("=" * 60)
    
    for rate in rates:
        mutated, _ = mutate(original, rate)
        original_prot = translate(original)
        mutated_prot = translate(mutated)
        
        prot_diff = sum(1 for a, b in zip(original_prot, mutated_prot) if a != b)
        
        print(f"\n{rate} mutation(s):")
        print(f"  Mutated: {mutated}")
        print(f"  Protein: {mutated_prot}")
        print(f"  Protein changes: {prot_diff}")

test_mutation_rates()
````

#### ðŸ”’ Preventing Stop Codon Mutations

````python
def mutate_safe(seq, n):
    """
    Mutate but avoid creating stop codons.
    """
    STOP_CODONS = ["TAA", "TAG", "TGA"]
    seq_list = list(seq)
    changes = []
    attempts = 0
    max_attempts = n * 100  # Prevent infinite loops
    
    while len(changes) < n and attempts < max_attempts:
        attempts += 1
        
        # Try a mutation
        idx = random.randrange(len(seq_list))
        old = seq_list[idx]
        new = random.choice([b for b in BASES if b != old])
        
        # Test if it creates a stop codon
        test_seq = seq_list.copy()
        test_seq[idx] = new
        
        # Check the codon this base is in
        codon_start = (idx // 3) * 3
        test_codon = "".join(test_seq[codon_start:codon_start+3])
        
        # Accept mutation if no stop codon created
        if test_codon not in STOP_CODONS:
            seq_list[idx] = new
            changes.append((idx, old, new))
    
    return "".join(seq_list), changes

# Test
original = "ATGCCGATTGCAGCG"
print(f"Original: {original}")
print(f"Protein:  {translate(original)}\n")

for i in range(3):
    mutated, changes = mutate_safe(original, 5)
    protein = translate(mutated)
    print(f"Try {i+1}: {mutated}")
    print(f"        {protein}\n")
````
