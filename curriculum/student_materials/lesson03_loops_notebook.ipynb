#### ðŸ”¬ Why Does GC Content Matter?

**GC Content** = Percentage of bases that are Guanine (G) or Cytosine (C)

#### Biological Significance:
1. **DNA Stability** - GC bonds are stronger (3 hydrogen bonds vs 2 for AT)
2. **Gene Density** - Higher GC often means more genes
3. **Species Identification** - Different organisms have characteristic GC%
4. **PCR Design** - Affects primer melting temperature
5. **Genome Features** - Promoters often have different GC than coding regions

#### Typical GC Content:
- *E. coli* (bacteria): ~50%
- Human genome: ~40%
- *Plasmodium* (malaria): ~20%

#### ðŸ”„ Introduction to For Loops

A **for loop** lets you do something for each item in a sequence.

#### For Loop Syntax

```python
for variable in sequence:
    # code to repeat
    # code block must be indented
```

**Key parts:**
- `for` - keyword that starts the loop
- `variable` - temporary name for each item (you choose this name)
- `in` - keyword that connects variable to sequence
- `sequence` - the string, list, or collection to iterate through
- `:` - colon marks the start of the code block
- **Indentation** - code inside the loop MUST be indented (4 spaces or 1 tab)

#### Example - Print Each Base

````python
# Simple loop - print each base
dna = "ATGCGT"

print("Each base in the sequence:")
for base in dna:
    print(base)  # This line is indented - it's inside the loop
````

#### ðŸ”¢ Counting with Loops

Let's count how many times a specific base appears:

````python
# Count G bases
dna = "ATGCCGATTA"
count_g = 0

for base in dna:
    if base == "G":
        count_g += 1  # Same as: count_g = count_g + 1

print(f"Sequence: {dna}")
print(f"Number of G: {count_g}")
````

#### ðŸ“Š Calculating GC Content

Formula: **GC% = (G + C) / Total Length Ã— 100**

````python
# Calculate GC content
dna = "ATGCCGATTA"

count_g = 0
count_c = 0

# Count G and C
for base in dna:
    if base == "G":
        count_g += 1
    if base == "C":
        count_c += 1

# Calculate percentage
gc_content = (count_g + count_c) / len(dna) * 100

print(f"Sequence: {dna}")
print(f"Length: {len(dna)} bases")
print(f"G count: {count_g}")
print(f"C count: {count_c}")
print(f"GC content: {gc_content:.1f}%")
````

#### ðŸŽ¨ Better Version - Using `or`

````python
# More elegant approach
dna = "ATGCCGATTA"
gc_count = 0

for base in dna:
    if base == "G" or base == "C":
        gc_count += 1

gc_content = gc_count / len(dna) * 100

print(f"Sequence: {dna}")
print(f"GC count: {gc_count}")
print(f"GC content: {gc_content:.1f}%")
````

#### ðŸ† Even Better - Using `in`

````python
# Most Pythonic approach
dna = "ATGCCGATTA"
gc_count = 0

for base in dna:
    if base in "GC":  # Checks if base is G or C
        gc_count += 1

gc_content = gc_count / len(dna) * 100

print(f"Sequence: {dna}")
print(f"GC%: {gc_content:.2f}%")
````

#### ðŸ”€ Conditional Statements: if, elif, else

Before we do complete frequency analysis, let's understand conditional statements.

#### Syntax

```python
if condition1:
    # code runs if condition1 is True
elif condition2:
    # code runs if condition1 is False AND condition2 is True
elif condition3:
    # code runs if previous conditions are False AND condition3 is True
else:
    # code runs if ALL previous conditions are False
```

**Key points:**
- `if` - checks the first condition
- `elif` - short for "else if", checks additional conditions (only if previous ones were False)
- `else` - catches everything else (optional)
- Only ONE block of code runs (the first True condition)
- Each condition needs a `:` at the end
- Code blocks must be indented

#### Example - Classify a Base

````python
# Classify a single nucleotide
base = "G"

if base == "A":
    print(f"{base} is Adenine (Purine)")
elif base == "G":
    print(f"{base} is Guanine (Purine)")
elif base == "C":
    print(f"{base} is Cytosine (Pyrimidine)")
elif base == "T":
    print(f"{base} is Thymine (Pyrimidine)")
else:
    print(f"{base} is not a valid DNA base!")

# Try changing base to "A", "X", etc.
````

#### Why Use elif Instead of Multiple if Statements?

**Using elif (efficient - stops after first match):**
```python
if base == "A":
    count_a += 1
elif base == "T":    # Only checked if base != "A"
    count_t += 1
elif base == "G":    # Only checked if base != "A" and base != "T"
    count_g += 1
```

**Using multiple ifs (inefficient - checks everything):**
```python
if base == "A":
    count_a += 1
if base == "T":      # Always checked, even if base == "A"
    count_t += 1
if base == "G":      # Always checked, even if already matched
    count_g += 1
```

With `elif`, once a condition is True, Python skips the rest!

#### ðŸ“ˆ Complete Nucleotide Frequency Analysis

Now let's use `elif` to count all four bases:

````python
# Count all four bases
dna = "ATGCCGATTA"

count_a = 0
count_t = 0
count_g = 0
count_c = 0

for base in dna:
    if base == "A":
        count_a += 1
    elif base == "T":
        count_t += 1
    elif base == "G":
        count_g += 1
    elif base == "C":
        count_c += 1

# Calculate percentages
total = len(dna)
print(f"Sequence: {dna}")
print(f"Total length: {total} bases\n")

print("Nucleotide Frequencies:")
print(f"  A: {count_a} ({count_a/total*100:.1f}%)")
print(f"  T: {count_t} ({count_t/total*100:.1f}%)")
print(f"  G: {count_g} ({count_g/total*100:.1f}%)")
print(f"  C: {count_c} ({count_c/total*100:.1f}%)")

# Verify totals
print(f"\nVerification: {count_a + count_t + count_g + count_c} = {total} âœ“")

# GC content
gc = (count_g + count_c) / total * 100
print(f"\nGC Content: {gc:.1f}%")
````

#### ðŸ“š Using Dictionaries for Frequency Counting

A cleaner way to store frequencies:

````python
# Count with dictionary
dna = "ATGCCGATTA"

# Initialize frequency dictionary
frequencies = {"A": 0, "T": 0, "G": 0, "C": 0}

# Count each base
for base in dna:
    if base in frequencies:
        frequencies[base] += 1

# Print results
print(f"Sequence: {dna}")
print(f"Length: {len(dna)} bases\n")

print("Frequency Dictionary:")
print(frequencies)

print("\nFormatted Report:")
for base, count in frequencies.items():
    percentage = count / len(dna) * 100
    print(f"  {base}: {count:2d} ({percentage:5.1f}%)")

# Calculate GC
gc_percent = (frequencies["G"] + frequencies["C"]) / len(dna) * 100
print(f"\nGC Content: {gc_percent:.1f}%")
````

#### ðŸ§¬ Real-World Example: Analyzing Different Genomes

#### ðŸ”¬ Advanced: Sliding Window GC Content

Analyze GC content in different regions of a sequence:

````python
# GC content in windows
dna = "ATATATGCGCGCGCTATATGCGCGCATATAT"
window_size = 10

print(f"Sequence: {dna}")
print(f"Analyzing {window_size}-base windows:\n")

for i in range(0, len(dna) - window_size + 1, 5):
    window = dna[i:i+window_size]
    
    gc_count = 0
    for base in window:
        if base in "GC":
            gc_count += 1
    
    gc_percent = gc_count / window_size * 100
    
    print(f"Position {i:2d}-{i+window_size:2d}: {window} | GC: {gc_percent:.0f}%")
````
