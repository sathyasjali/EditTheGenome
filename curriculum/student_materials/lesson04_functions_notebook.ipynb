##### ðŸ¤” Why Use Functions?

##### Without Functions (Copy-Paste Hell):
```python
# Analyzing sequence 1
seq1 = "ATGC..."
gc_count = 0
for base in seq1:
    if base in "GC":
        gc_count += 1
gc1 = gc_count / len(seq1) * 100

# Analyzing sequence 2 - COPY-PASTE!
seq2 = "GCTA..."
gc_count = 0
for base in seq2:
    if base in "GC":
        gc_count += 1
gc2 = gc_count / len(seq2) * 100
```

##### With Functions (Reusable!):
```python
def gc_content(seq):
    gc_count = sum(1 for base in seq if base in "GC")
    return gc_count / len(seq) * 100

gc1 = gc_content("ATGC...")
gc2 = gc_content("GCTA...")
```

**Benefits:** Less code, fewer bugs, easier to maintain!


##### ðŸŽ¯ Understanding Functions Deeply

**What IS a Function?**

A function is a named block of reusable code that:
1. **Takes input** (parameters)
2. **Does something** (processes)
3. **Returns output** (result)

**Terminology:**

| Term | Meaning | Example |
|------|---------|---------|  
| Define | Create function | `def my_func():` |
| Call | Use function | `my_func()` |
| Parameter | Variable in definition | `def func(param):` |
| Argument | Actual value passed | `func("ATGC")` |
| Return | Give back result | `return result` |



````python
# Anatomy of a Function

def calculate_gc(sequence):
    """
    Calculate GC content.
    
    Args:
        sequence (str): DNA
    
    Returns:
        float: GC percentage
    """
    g = sequence.count("G")
    c = sequence.count("C")
    total = len(sequence)
    gc = (g + c) / total * 100
    return gc

# Calling the function
dna = "ATGCCGATTA"
result = calculate_gc(dna)
print(f"GC%: {result:.1f}")
````

##### ðŸ“¥ Parameters vs Arguments

**Parameter:** Variable in function definition
```python
def greet(name):  # â† 'name' is parameter
    return f"Hello, {name}"
```

**Argument:** Actual value when calling
```python
greet("Alice")    # â† "Alice" is argument
```

**Think of it as:**
- Parameter = blank on a form: "Name: _____"
- Argument = filled value: "Name: Alice"



````python
# Parameters vs Arguments

def compare_sequences(seq1, seq2):
    """Compare length of two sequences."""
    len1 = len(seq1)
    len2 = len(seq2)
    
    if len1 > len2:
        return f"{seq1} is longer"
    elif len2 > len1:
        return f"{seq2} is longer"
    else:
        return "Same length"

# Positional arguments
dna1 = "ATGC"
dna2 = "ATGCGT"
print(compare_sequences(dna1, dna2))

print()

# Keyword arguments
print(compare_sequences(seq2=dna2, seq1=dna1))
````

##### ðŸ”™ Understanding `return`

**What does `return` do?**
1. **Ends function immediately** - no code after runs
2. **Sends value back** to caller
3. **Optional** - without return, gives back `None`

**With return:**
```python
def gc(seq):
    result = (seq.count("G") + seq.count("C")) / len(seq) * 100
    return result  # Sends back

value = gc("ATGC")  # Gets result
print(value)        # Can use it
```

**Without return:**
```python
def print_gc(seq):
    result = (seq.count("G") + seq.count("C")) / len(seq) * 100
    print(result)   # Just prints

value = print_gc("ATGC")  # Prints 50.0
print(value)              # Prints None
```



````python
# Understanding return

# WITH return
def add(a, b):
    result = a + b
    return result
    print("Never prints")  # After return = unreachable

sum_val = add(5, 3)
print(f"Sum: {sum_val}")

print("\n" + "="*50 + "\n")

# WITHOUT return
def just_print(text):
    print(text)
    # No return

result = just_print("Hello")
print(f"Returned: {result}")  # None

print("\n" + "="*50 + "\n")

# Early return
def validate_dna(seq):
    for base in seq.upper():
        if base not in "ATGC":
            return False  # Exit immediately
    return True  # Only if all valid

print(f"'ATGC' valid: {validate_dna('ATGC')}")
print(f"'ATGCX' valid: {validate_dna('ATGCX')}")
````
